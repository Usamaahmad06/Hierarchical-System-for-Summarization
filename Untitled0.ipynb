```python
import torch
from torch import nn, optim
from torchvision import transforms, datasets
from avalanche.benchmarks.classic import SplitMNIST
from avalanche.models import SimpleMLP
from avalanche.training.supervised import Naive
from avalanche.evaluation.metrics import accuracy_metrics, loss_metrics, timing_metrics
from avalanche.logging import InteractiveLogger
from avalanche.training.plugins import EvaluationPlugin

# Define the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define the benchmark - SplitMNIST automatically downloads and splits the MNIST dataset
benchmark = SplitMNIST(n_experiences=5, seed=1)

# Define the model
model = SimpleMLP(num_classes=10)

# Define the optimizer and loss function
optimizer = optim.SGD(model.parameters(), lr=0.01)
criterion = nn.CrossEntropyLoss()

# Define the evaluation plugin
interactive_logger = InteractiveLogger()
evaluation_plugin = EvaluationPlugin(
    accuracy_metrics(minibatch=True, epoch=True, experience=True, stream=True),
    loss_metrics(minibatch=True, epoch=True, experience=True, stream=True),
    timing_metrics(epoch=True),
    loggers=[interactive_logger]
)

# Define the strategy
cl_strategy = Naive(
    model, optimizer, criterion,
    train_mb_size=32, train_epochs=2, eval_mb_size=32,
    device=device, evaluator=evaluation_plugin
)

# Training loop
for experience in benchmark.train_stream:
    print("Start training on experience ", experience.current_experience)
    cl_strategy.train(experience)  # Train on the current experience's dataset
    print("End training on experience ", experience.current_experience)
    print("Computing accuracy on the test set")
    cl_strategy.eval(benchmark.test_stream)  # Evaluate on the test set
```

```python
import torch
from torchvision import transforms
from avalanche.benchmarks.classic import SplitMNIST
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt  # Import matplotlib for visualization

# Define the benchmark - SplitMNIST automatically downloads and splits the MNIST dataset
benchmark = SplitMNIST(n_experiences=5, seed=1, return_task_id=True)

# Access the training dataset from the first experience
train_dataset = benchmark.train_stream[0].dataset

# Create a DataLoader to iterate over the dataset
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=False)

# Define a transform to visualize the images
transform = transforms.ToPILImage()

# Print the first 15 instances
print("First 15 instances of the SplitMNIST dataset:")
for idx, (data, target, task_id) in enumerate(train_loader):
    if idx >= 15:
        break
    image = data.squeeze(0)
    label = target.item()
    task = task_id.item()

    # Convert tensor to image for display purposes
    img = transform(image)
    img.show()

    # Print the instance information
    print(f"Instance {idx + 1}: Label = {label}, Task ID = {task}")

    # Display the image using matplotlib
    plt.imshow(image.squeeze(0), cmap='gray')
    plt.title(f"Label: {label}, Task ID: {task}")
    plt.show()
```

```python
import torch
from torch.nn import CrossEntropyLoss
from torch.optim import SGD
from torch.utils.data import DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt

from avalanche.benchmarks.classic import PermutedMNIST
from avalanche.models import SimpleMLP
from avalanche.training import Naive

# Config
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Model
model = SimpleMLP(num_classes=10)

# CL Benchmark Creation
perm_mnist = PermutedMNIST(n_experiences=3)
train_stream = perm_mnist.train_stream
test_stream = perm_mnist.test_stream

# Prepare for training & testing
optimizer = SGD(model.parameters(), lr=0.001, momentum=0.9)
criterion = CrossEntropyLoss()

# Continual learning strategy
cl_strategy = Naive(
    model, optimizer, criterion, train_mb_size=32, train_epochs=2,
    eval_mb_size=32, device=device
)

# Print the first 10 instances from the first training experience
print("First 10 instances of the PermutedMNIST dataset:")

# Get the first experience
first_experience = train_stream[0].dataset

# Create a DataLoader for the first experience dataset
data_loader = DataLoader(first_experience, batch_size=1, shuffle=False)

# Transform to display images
transform = transforms.ToPILImage()

# Loop through the first 10 instances and display them
for idx, batch in enumerate(data_loader):
    if idx >= 10:
        break
    data, target = batch[:2]
    image = data.squeeze(0)
    label = target.item()

    # Convert tensor to image for display purposes
    img = transform(image)

    # Display the image using matplotlib
    plt.imshow(image.squeeze(0), cmap='gray')
    plt.title(f"Label: {label}")
    plt.show()

    # Print the instance information
    print(f"Instance {idx + 1}: Label = {label}")

# Train and test loop over the stream of experiences
results = []
for train_exp in train_stream:
    cl_strategy.train(train_exp)
    results.append(cl_strategy.eval(test_stream))
```
